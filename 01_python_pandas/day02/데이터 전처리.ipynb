```python
add_col['사회'] = Series([82,90,592,64,77], index=['철수','영희','민수','수현','민철'])
add_col
```

```python
# conditions의 리스트 순서에 맞게 학점 부여
grade = ['A','B','C','F', 'F']

add_col['영어학점'] = numpy.select(conditions,grade)
add_col
```

### 2) 열 단위 병합
#### 한 번에 두 개의 데이터 프레임만 병합 가능

#### a) 인덱스가 지정하지 않는 샘플 데이터 프레임 만들기

```python
df_left = DataFrame({'고객번호':[1001,1002,1003,1004],'이름':['철수','영희','민철','미영']})
df_left
```

```python
df_right = DataFrame({'고객번호':[1001,1002,1003,1005],'금액':[10000,20000,15000,5000]})
df_right
```

#### b) 두 데이터 프레임의 공통 컬럼을 기준으로 병합하기
양쪽 데이터 프레임에 모두 존재하는 데이터만 보여준다.
> DataBase의 inner join과 같은 개념

```python
# 병합 시 제외되는 항목들
# -df_left의 고객번호 1004번 데이터는 df_right의 고객번호 컬럼과 겹치지 않는다.
# -df_right의 고객번호 1004번 데이터는 df_left의 고객번호 컬럼과 겹치지 않는다.
merge(df_left,df_right)
```

#### c) 데이터 프레임을 기준을 정해서 병합하기
`how` 파라미터는 병합할 기준이 되는 데이터프레임을 정한다. <br>
`how= left` : 왼쪽 데이터프레임의 모든 데이터<br>
`how = right` : 오른쪽 데이터프레임의 모든 데이터<br>
`how = outer` : 공통 컬럼의 데이터가 한쪽에만 있어도 데이터프레임의 모든 데이터를 보여준다.<br>
how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
> DataBase의 Left Join, Right Join, Full Outer Join ,cross join

```python
merge(df_left, df_right, how="left")
```

```python
merge(df_left, df_right, how="right")
```

```python
merge(df_left, df_right, how="outer")
```

#### f) 중복되는 데이터가 존재하는 경우 열단위 병합

```python
df_first = DataFrame({'아이디':['hello','world','python','hello'], '결제금액' : [14000,13000,15000,13000]})
df_first
```

```python
df_second = DataFrame({'아이디':['hello','python','python','world'], '결제금액' : [300,500,100,200]})
df_second
```

```python
# on = '아이디' : 아이디 명을 기준으로 병합한다.
merge(df_first, df_second, on='아이디')
```
