## #01. 필요한 패키지 및 샘플 데이터 준비

### 1) 패키지 참조

````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
````

### 데이터 가져오기

> 데이터 출처 : [Kosis 국가 통계 포털] [https://kosis.kr/index/index.do]

````python
교통사고csv = pd.read_csv('traffic.csv',encoding='euc-kr')
교통사고csv
````

### 3) 데이터 전처리 (년도별 합계 구하기)

````python
교통사고tmp = 교통사고csv.filter(['년도','발생건수','사망자수','부상자수'])
교통사고df = 교통사고tmp.groupby(['년도']).sum()
교통사고df
````

## #04. 세로 막대 그래프
### 1) 특정 컬럼에 대한 시각화
#### 기본 옵션만 사용하기 
```python
    데이터프레임객체[ '컬럼' ].plot.bar()
```
* rot 파라미터는  x 축 텍스트의 각도, 기본값 90
* width 파라미터는 모든 막대가 공백없이 꽉 찬 상태를 1로 하여 비율적으로 지정, 기본값은 0.5

막대그래프는 index의 데이터 타입이 숫자, 문자열에 상관없이 x축이 0부터 시작되는 좌표계를 갖는다.

그러므로 `xticks()` 함수를 사용할 때 반드시 x 축 좌표에 대한 지정이 이루어져야 한다.

````python
교통사고df['발생건수'].plot.bar(color = 'red', rot=0)
xpos = np.arange(len(교통사고df.index))

plt.grid()
plt.legend() #범례
plt.title('년도별 교통사고 발생 현황')
plt.xlabel("년도")
plt.ylabel('발생 건수')
# 만들어 놓은 xtext 넣기 
plt.xticks(xpos, xtext)
plt.show()
````

####  텍스트 출력 처리, Y축 범위 지정

#### 텍스트 출력은 그래프의 x,y좌표를 구해서

````python
교통사고df['발생건수'].plot.bar(color = 'red', width = 0.7, rot=0)
xpos = np.arange(len(교통사고df.index))
plt.grid()
plt.legend() #범례
plt.title('년도별 교통사고 발생 현황')
plt.xlabel("년도")
plt.ylabel('발생 건수')
# 만들어 놓은 xtext 넣기 
plt.xticks(xpos, xtext)

# y축의 범위 설정 / x축의 범위 설정 (xlim(min, max))사용
plt.ylim(200000, 250000)

# 그래프에 텍스트 표시하기
# 그래프에 표시되는 실 데이터는 y축 좌표를 의미하고, x축은 0부터 시작하는 좌표값을 갖는다..
for k,v in enumerate(list(교통사고df['발생건수'])):
    txt = '%d건' %v
    plt.text(k,v,txt, fontsize = 14, color = '#0000ff', horizontalalignment = "center", verticalalignment = "bottom")
plt.show()
````

````python
교통사고df.plot.bar(width=0.7, rot=0)
xpos = np.arange(len(교통사고df.index))

plt.grid()
#plt.legend() #범례
# 범례 위치 변경(x좌표, y좌표)
# plt.legend(bbox_to_anchor=(0,0))
plt.legend(bbox_to_anchor=(1,0.5))

plt.title('년도별 교통사고 발생 현황')
plt.xlabel("년도")
plt.ylabel('발생 건수')
# 만들어 놓은 xtext 넣기 
plt.xticks(xpos, xtext)
plt.show()
````

## #06. 파이 그래프

데이터프레임의 특정 컬럼을 지정하여 전체를 기준으로 어느정도의 비율을 차지하는지 시각화하는 그래프 

### 1) 파이 그래프 기본형

````python
# 각 컬럼별로 평균 구하기
교통사고평균 = 교통사고df.mean()
교통사고평균
````

````python
# 구해진 결과를 가지고 데이터프레임 생성
교통사고평균df = pd.DataFrame(교통사고평균, columns=['교통사고'])
교통사고평균df
````

#### 파이크 그래프 표시 -> 반드시 특정 컬럼을 지정해야 한다.

````python
교통사고평균df['교통사고'].plot.pie()
plt.show()
````

### 1) 파이 그래프의 옵션

````python
# 4. 건수를 리턴하는 함수 만들기
def make_autopct(비율):
    per = '%0.1f%%' %비율
    total = 교통사고평균df['교통사고'].sum()
    cnt = int(total * 비율/100)
    return "{} \n ({} 건)".format(per, cnt)

# 파이 그래프 생성
교통사고평균df['교통사고'].plot.pie(labels=['전체','사망','부상'],
                            colors=['#ff6600','#0066ff','#ffff00'],
                            autopct=make_autopct, startangle=90)

# 1.타이틀
plt.title("교통사고비율")
# 2.y라벨 없애기
plt.ylabel(None)
# 3.범주
plt.legend(bbox_to_anchor=(1,0.5))

plt.show()
````

## #07. 산점도 그래프
- 두 변수간의 영향력을ㄹ 보여주기 위해ㅐ서 가로축과 세로축에 `데이터 포인트`를 그리는데 사용
- `데이터 포인트`들이 뭉쳐 있을 수록 두 변수간의 관련성 정도가 높다고 해석한다.

### 1) 교통사고 발생건수와 부상자 수간의 상관관계
- 첫번째 변수 : 발생건 수
- 두번째 변수 : 부상자 수

````python
# marker -> o, v, ^, < >, 8, s, p, *, h ,H, D, d, P, X
교통사고csv.plot.scatter(x='발생건수', y='부상자수', c='r',  marker='o')
plt.grid()
plt.title('발생건수과 부상자 수의 상관관계')
plt.ylabel('부상자 수')
plt.xlabel('발생건수')

plt.show()
````

### 산점도 그래프 해석

점들이 많이 모여 있을 수록 두 변수간의 관계썽이 높고, 점들이 흩어져 있을 술록 두 변수간의 관계성이 적다고 해석

### 정의 관계 : x가 증가할 때 y도 증가하는 직선에 가까운 형태.
### 역의 관계 : x가 증가할 때 y는 감소하는 형태.
### 선형 관계 : 점들이 직선에 가깝게 배치된 형태
### 비선형 관계 : 점들이 직선이 아닌 곡선 형태

### 상관분석

연속형 변수로 측정된 두 변수간의 선형적 관계를 분석하는 관계

연속형 변수는 산술 평균을 계산할 수 있는 숫자형의 데이터이며, 선형적 관계라 함은 흔히 비례식이 성립되는 관계를 말한다.

> 예를들어 A변수가 증가함에 따라 B변수도 증가되는지 혹은 감소하는지를 분석하는 것이다.

상관분석은 두 변수간의 선형적 관계에 초점을 두고 있으며 **선형관계를 갖는가? 어느방향으로 형성되어 있는가? 그 관계의 정도는 얼마나 큰가?**와 

같은 질문을 던지며 결과 해석을 수행한다.

### 상관계수

상관분석에서 두 변수 사이의 선형적인 관계정도를 나타내기 위해 사용되는 값.

> 수하적인 공식에 의해 구해야 한다.

상관분석에는 측정 데이터에 따라 피어슨 상관분석, 상곤분석 등의 여러가지 분석방법이 있지만, 일반적으로 **피어슨 상관계수**를 의미한다.

| 상관계수 | 설명 |
| -------- | ---- |
|  1  | 직선형 정의 관계 |
|  -1 | 직선형 역의 관계 |
|  0  | 모든 점들이 분산되어 있는 형태 |

> 상관계수 1과 0 사이에 소수점 형태로 점의 분포도를 표현한다.

### 추세선

차트 위에 그려진 대각선들

이는 특정 데이터 지점을 연결하며 분석가가 보다 쉽게 데이터의 움직임을 시각화하고 추세를 알 수 있게한다.

추세선은 기술적 분석의 가장 기초적인 도구 중 하나로 간주된다.

````python
# 추세선
x = 교통사고csv["발생건수"]
y = 교통사고csv["부상자수"]

# x,y를 기반으로 추세선을 위한 계산
z = np.polyfit(x,y,1) # (x,y, 1차원)
f = np.poly1d(z)
````

````python
교통사고csv.plot.scatter(x='발생건수', y='부상자수', c='r',  marker='o')
plt.grid()
plt.title('발생건수과 부상자 수의 상관관계')
plt.ylabel('부상자 수')
plt.xlabel('발생건수')
plt.plot(x,f(x),c='b')
plt.show()
````
